```
df_sales = spark.read.option("header", True).csv("dbfs:/FileStore/tables/ratings.csv")
df_sales.show()
```

```
df_sales.write.format("parquet").mode("overwrite").option("path","/data/output/sales_parquet/").saveAsTable("sales_parquet")
```

```
display(dbutils.fs.ls("/data/output/sales_parquet/"))
```
